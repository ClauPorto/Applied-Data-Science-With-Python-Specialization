## Step 1
Load the dataset and transform the data into Celsius (refer to documentation) then extract all of the rows which have minimum or maximum temperatures.

__hint: when I did this step I had two DataFrame objects, each with ~80,000 entries in it__

```python
import pandas as pd
df = pd.read_csv('assets/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv')
df.head()
```

```python
# In this code cell, transform the Data_Value column

# Transform Data_Value to Celsius (originally in tenths of degrees Celsius)
df['Data_Value'] = df['Data_Value'] / 10

# Convert 'Date' to datetime
df['Date'] =  pd.to_datetime(df['Date'], infer_datetime_format=True)

# Extract rows with minimum temperatures
df_tmin = df[df['Element'] == 'TMIN'].sort_values(by='ID').reset_index(drop=True)

# Extract rows with maximum temperatures
df_tmax = df[df['Element'] == 'TMAX'].sort_values(by='ID').reset_index(drop=True)
             
print(df_tmin.shape)
df_tmin.head()
```

```python
print(df_tmax.shape)
df_tmax.head() 
```

## Step 2
In order to visualize the data we would plot the min and max data for each day of the year between the years 2005 and 2014 across all weather stations. But we also need to find out when the min or max temperature in 2015 falls below the min or rises above the max for the previous decade.

If you did step 1 you have two Series objects with min and max times for the years 2005 through 2015. You can use Pandas `groupby` to create max and min temperature Series objects across all weather stations for each day of these years, and you can deal with the records for February 29 (the leap year) by dropping them.

__hint: when I finished this step, I had two DataFrame objects, each with exactly 4015 observations in them__

```python
# create a DataFrame of minimum and maximum temperatures by date

# Filter data for the years 2005 to 2014
df_tmin_decade = df_tmin[(df_tmin['Date'].dt.year >= 2005) & (df_tmin['Date'].dt.year <= 2014)]
df_tmax_decade = df_tmax[(df_tmax['Date'].dt.year >= 2005) & (df_tmax['Date'].dt.year <= 2014)]

# Add 'Day_Of_Year' column
df_tmin_decade['Day_Of_Year'] = df_tmin_decade['Date'].dt.dayofyear
df_tmax_decade['Day_Of_Year'] = df_tmax_decade['Date'].dt.dayofyear

# Add 'Month_Of_Year' column
df_tmin_decade['Month_Of_Year'] = df_tmin_decade['Date'].dt.month
df_tmax_decade['Month_Of_Year'] = df_tmax_decade['Date'].dt.month

# Remove February 29
df_tmin_decade = df_tmin_decade[~((df_tmin_decade['Date'].dt.day == 29) & (df_tmin_decade['Date'].dt.month == 2))]
df_tmax_decade = df_tmax_decade[~((df_tmax_decade['Date'].dt.day == 29) & (df_tmax_decade['Date'].dt.month == 2))]

# Group by 'Day_Of_Year' and calculate min and max temperatures for each day
df_tmin_decade = df_tmin_decade.groupby(by='Day_Of_Year')['Data_Value', 'Month_Of_Year'].min().reset_index()
df_tmax_decade = df_tmax_decade.groupby(by='Day_Of_Year')['Data_Value', 'Month_Of_Year'].max().reset_index()
```

## Step 3
Now that you have grouped the daily max and min temperatures for each day of the years 2005 through 2015, you can separate out the data for 2015. Then you can use the Pandas `groupby` function to find the max and min of the temperature data for each __day of the year__ for the 2005-2014 data.

__hint: at the end of this step I had two DataFrames, one of maximum and the other of minimum values, which each had 365 observations in them. I also had another pair of similar DataFrames but only for the year 2015.__

```python
# calculate the minimum and maximum values for the years 2015
# Filter data for the year 2015
df_tmin_2015 = df_tmin[df_tmin['Date'].dt.year == 2015]
df_tmax_2015 = df_tmax[df_tmax['Date'].dt.year == 2015]

# Add 'Day_Of_Year' column for 2015 data
df_tmin_2015['Day_Of_Year'] = df_tmin_2015['Date'].dt.dayofyear
df_tmax_2015['Day_Of_Year'] = df_tmax_2015['Date'].dt.dayofyear

# Add 'Month_Of_Year' column
df_tmin_2015['Month_Of_Year'] = df_tmin_2015['Date'].dt.month
df_tmax_2015['Month_Of_Year'] = df_tmax_2015['Date'].dt.month

# Remove February 29
df_tmin_2015 = df_tmin_2015[~((df_tmin_2015['Date'].dt.day == 29) & (df_tmin_2015['Date'].dt.month == 2))]
df_tmax_2015 = df_tmax_2015[~((df_tmax_2015['Date'].dt.day == 29) & (df_tmax_2015['Date'].dt.month == 2))]

# Group by 'Day_Of_Year' and calculate min and max temperatures for each day in 2015
df_tmin_2015 = df_tmin_2015.groupby('Day_Of_Year')['Data_Value', 'Month_Of_Year'].min().reset_index()
df_tmax_2015 = df_tmax_2015.groupby('Day_Of_Year')['Data_Value', 'Month_Of_Year'].max().reset_index()

# Merge the 2015 data with the decade data to compare
comparison_tmin = pd.merge(df_tmin_decade, df_tmin_2015, on='Day_Of_Year', suffixes=('_decade', '_2015'))
comparison_tmax = pd.merge(df_tmax_decade, df_tmax_2015, on='Day_Of_Year', suffixes=('_decade', '_2015'))

# Identify records where 2015 min is less than the decade min and 2015 max is greater than the decade max
record_low_2015 = comparison_tmin[comparison_tmin['Data_Value_2015'] < comparison_tmin['Data_Value_decade']]
record_high_2015 = comparison_tmax[comparison_tmax['Data_Value_2015'] > comparison_tmax['Data_Value_decade']]
```

## Step 4
Now it's time to plot! You need to explore matplotlib in order to plot line graphs of the min and max temperatures for the years 2005 through 2014 and to scatter plot __only__ the daily 2015 temperatures that exceeded those values.

```python
import matplotlib.pyplot as plt
from calendar import month_abbr

# put your plotting code here!

plt.figure(figsize=(18,13))
xticks = pd.date_range('2015-01-01', '2015-12-31', freq='M').strftime('%j').astype(int)
xticks_labels = pd.to_datetime(xticks, format='%j').strftime('%b')

yticks = list(range(-40,50, 10))

plt.plot(df_tmin_decade['Day_Of_Year'], df_tmin_decade['Data_Value'], color="blue", alpha=0.5, label="Min (2005-2014)")
plt.plot(df_tmax_decade['Day_Of_Year'], df_tmax_decade['Data_Value'], color="red", alpha=0.5, label="Max (2005-2014)")
plt.fill_between(df_tmin_decade['Day_Of_Year'], df_tmin_decade['Data_Value'], df_tmax_decade['Data_Value'], color='grey', alpha=0.2)

plt.scatter(record_low_2015['Day_Of_Year'], record_low_2015['Data_Value_2015'], color="black", label="Record Breaking (2015)")
plt.scatter(record_high_2015['Day_Of_Year'], record_high_2015['Data_Value_2015'], color="black")

plt.xlabel("Month", fontsize="xx-large")
plt.ylabel("Temperature °C", fontsize="xx-large")
plt.title("Min and Max temperatures for the years 2005-2014 vs Record Breaking of 2015", fontsize="xx-large")
plt.xticks(xticks, xticks_labels, fontsize="xx-large")
plt.yticks(yticks, fontsize="xx-large")

# remove the frame of the chart
for spine in plt.gca().spines.values():
    spine.set_visible(False)

# Increase the font size of the legend
plt.legend(fontsize="xx-large")

# Save the figure before showing it
plt.savefig("assignment.jpg")

plt.show()
```
